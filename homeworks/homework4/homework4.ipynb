## Задание 1

Используя встроенные функции numpy (np.random.normal) сгенирируйте нормальное распределение c размерами выборки 10, 100, 1000 и 10000. Постройте гистограммы распределений. Рассчитайте среднее значение, медиану и моду для каждой выборке и сравните эти величины в пределах выборки.

### Решение:

```python
def plot_vertical(plt, x, y, ls, c):
    plt.plot([x,x], [0,y], ls=ls, c=c)
```

n=10

```python
n = 10

samples = np.random.normal(size=n)

mean = np.mean(samples)
median = np.median(samples)
mode = stats.mode(np.round(samples, 1))[0]


print('Среднее: ', mean)
print('Медиана: ', median)
print('Мода: ', mode)
```

```python
y_max = 4
plt.hist(samples)

plot_vertical(plt, x=mean, y=y_max, ls='--', c='red')
plot_vertical(plt, x=median, y=y_max, ls='--', c='green')
plot_vertical(plt, x=mode, y=y_max, ls='--', c='orange')

plt.legend({'mean':'red', 'median':'green', 'mode': 'orange'})
plt.title('Расределение вероятнотей n=' + str(n))
plt.show()
```

## Задание 2

Плотность распределения вероятностей нормальной случайной величины X имеет вид f(x) = g e^{-x^2 + 6x + 3}. Требуется найти: a) неизвестный параметр g b) математическое ожидание M(X) и дисперсию D(X)

### Решение:

Формула плотности нормального распределеня имеет вид:

$$ \large f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}},$$

Наша задача свернуть выражение, расположенние в степени экспоненты, до полного квадрата и выделить знаменатель и -1. Тогда мы легко найдет M(X)  и D(X). А затем легко вычислим параметр g

Преобразуем сначала выражение:

$$ \large e^{x^2+6x+3} = e^{-(x^2-6x+9)+12}= e^{12}\cdot e^{-\frac{(x-3)^2}{2\cdot 0.5}}$$

Отсюда:

$$ \large D(X) = \sigma^2 = 0.5 $$

$$ \large M(X) = a = 3 $$

$$ \large g \cdot e^{12} = \frac{1}{0.5 \cdot \sqrt{2\pi}} ; \\ g = \frac{1}{0.5 e^{12}\sqrt{2\pi}} $$

Выпишем конечный результат

$$ \large f(x) = e^{12} \cdot \frac{1}{0.5 e^{12} \sqrt{2\pi}} \cdot e^{-\frac{(x-3)^2}{2\cdot 0.5}} $$
