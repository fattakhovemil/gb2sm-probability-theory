### Задание 4
В первой урне содержится 10 шаров, из них 8 белых; во второй урне 20 шаров, из них 4 белых. Из каждой урны наудачу извлекли по одному шару, а затем из этих двух шаров наудачу взят один шар. Найти вероятность того, что взят белый шар.

#### Решение:



Разложим наше событие на несколько более простых событий, для этого рассмотрим несколько возможных сценариев:
 1) Событие A, когда из обеих урн сразу достали два белых шара. Дальше не важно, какой шар будет взят, любой из них будет белый
 2) Событие B, когда из одной из урн достали белый шар, а из другой урны достали черный шар. И следом зависимое Событие C, когда из черного и белого шара выбрали белый.
 
 Вероятность итогового события равна: $$ P = P(A) + P(B*C)  $$

__1.Найдем вероятность P(A).__ Она равна $$ P(A) = P(A1) * P(A2) $$ 

где A1 - событие, что из первой урны достали черный шар, A2 - событие, что из второй урны достали черный шар. События независимые, поэтому можем перемножать для нахождения вероятности совместного события

$$ P(A) = \frac{8}{10} \cdot \frac{4}{20} = \frac{4}{25} $$

__2.Найдем вероятность P(B).__

Событие B можно разложить на два более простых события: B1, когда из первой урны достали черный шар и одновременно из второй урны достали белый шар, и событие B2, когда из первой урны достали белый шар и одновременно из второй урны достали черный шар. Достаточно, чтобы произошло любое из этих событий. События B1 и B2 в свою очередь расчитываются аналогично событию A.

$$ P(B) = P(B1+B2) = P(B1) + P(B2) = \frac{2 \ черных}{10} \cdot \frac{4 \ белых}{20} + \frac{8 \ белых}{10} \cdot \frac{16 \ черных}{20} = \frac{1}{25} + \frac{16}{25} = \frac{17}{25} $$

__3.Найдем вероятность P(B * C).__

Событие С происходит только, если произошло событие B. Это зависимое событие. Поэтому вероятность находим по формуле:

$$ P(B*C) = P(B) * P(C|B) = \frac{17}{25} \cdot \frac{1}{2} = \frac{17}{50}  $$

__4.Найдем итоговую вероятность P.__

Для нас достаточно, чтобы произошло или событие А или зависимые события B и С.


$$ P = P(A) + P(B*C) = \frac{4}{25} + \frac{17}{50} = \frac{1}{2} $$

### Задание 5

Три стрелка произвели залп, причём две пули поразили мишень. Найти вероятность того, что третий стрелок поразил мишень, если вероятности попадания в мишень первым, вторым и третьим стрелками соответственно равны 0.6, 0.5 и 0.4.

#### Решение:



Обозначим следующие события:
 - А - два любых стрелка поразили мишень
 - $ S_1, S_2, S_3 $ - поражение мишени каждым из стрелков (вероятности известны по условию)
 - $ \neg{S_1},\neg{S_2},\neg{S_3} $ - промах по мишени каждым из стрелков (вероятности находятся как 1 - $ S_i $)

Искомая вероятность по формуле Байсе равна:

$$ P(S_3|A) = \frac{P(S_3) \cdot P(A|S_3)}{P(A)} $$

Вероятность $P(S_3)= 0.4$ 

Найдем $P(A|S_3)$. Это условная вероятность события А, когда точно известно, что третий стрелок поразил мишень.

Для этого необходимо, чтобы первый стрелок попал а второй промазал или первый промазал, а второй попал.

$$ P(A|S_3) = P(S_1) * P(\neg{S_2}) + P(\neg{S_1}) * P(S_2) = $$ 
$$ = 0.6 * (1-0.5) + (1-0.6) * 0.5 = 0.5 $$

Найдем  вероятность события, что 2 любых стрелка попали. Для этого мы можем разложить это события на два более простых, когда два стрелка попали, когда третий точно попал и когда 2 стрелка попали, когда третий точно промазал

$$ P(A) = P(S_3) * P(A|S_3) + P(\neg{S_3}) * P(A|\neg{S_3})  $$

В этом выражении нам не известно только $ P(A|\neg{S_3}) $

Если третий стрелок точно промазал, то два первых стрелка должны одновременно попасть в мишень. Это независимые события:

$$ P(A|\neg{S_3}) = P(S_1) * P(S_2) = 0.6 * 0.5 = 0.3 $$

Теперь вычислим P(A):

$$ P(A) = 0.4 * 0.5 + (1 - 0.4) * 0.3 = 0.38  $$

Подставим все в итоговую формулу:

$$ P(S_3|A) = \frac{0.4 \cdot 0.5}{0.38} = \frac{10}{19} = 0.526... $$
