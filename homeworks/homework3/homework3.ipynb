### Решение

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
plt.style.available
```

```python
df = pd.read_csv('../../lessons/lesson3/global_power_plant_database.csv', ';')
df.head()
```

```python
df.info()
```

#### capacity_mw

```python
print('Среднее значение:', df['capacity_mw'].mean())
print('Мода:', df['capacity_mw'].mode()[0])
print('Медиана:', df['capacity_mw'].median())
print('Среднеквадратичное отклонение:', df['capacity_mw'].std())
print('Смещенная дисперсия:', df['capacity_mw'].var(ddof=0))
print('Несмещенная дисперсия:', df['capacity_mw'].var(ddof=1))
print('1-й квартиль:', df['capacity_mw'].quantile(.25))
print('2-й квартиль:', df['capacity_mw'].quantile(.5))
print('3-й квартиль:', df['capacity_mw'].quantile(.75))
```

```python
plt.hist(df['capacity_mw'], bins=20)
plt.title("Распределение capacity_mw")
plt.xlabel("capacity_mw")
plt.ylabel("Количество")
plt.show()
```

```python
sns.boxplot(df['capacity_mw'], orient='h')
plt.title("Boxplot capacity_mw")
plt.show()
```

#### estimated_generation_gwh

```python
print('Среднее значение:', df['estimated_generation_gwh'].mean())
print('Мода:', df['estimated_generation_gwh'].mode()[0])
print('Медиана:', df['estimated_generation_gwh'].median())
print('Среднеквадратичное отклонение:', df['estimated_generation_gwh'].std())
print('Смещенная дисперсия:', df['estimated_generation_gwh'].var(ddof=0))
print('Несмещенная дисперсия:', df['estimated_generation_gwh'].var(ddof=1))
print('1-й квартиль:', df['estimated_generation_gwh'].quantile(.25))
print('2-й квартиль:', df['estimated_generation_gwh'].quantile(.5))
print('3-й квартиль:', df['estimated_generation_gwh'].quantile(.75))
```

```python
plt.hist(df['estimated_generation_gwh'])
plt.title("Распределение estimated_generation_gwh")
plt.xlabel("estimated_generation_gwh")
plt.ylabel("Количество")
plt.show()
```

```python
sns.boxplot(df['estimated_generation_gwh'], orient='h')
plt.title("Boxplot estimated_generation_gwh")
plt.show()
```
